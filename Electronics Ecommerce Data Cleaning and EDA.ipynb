```python
# Exploratory Data Analysis (Group Data and Visualize)


#Group data  (Product and Order totals) 5 highest sales products
df.groupby('Product').sum('Order Total')

# The top 5 highest sales products are: Macbook Pro Laptop, Iphone, ThinkPad Laptop, Google Phone,34in Ultrawide Monitor
```

```python
# Top 4 Cities with the highest and top 4 lowest total sales.
df.groupby("City").sum("Order Total")

# Top 4 highest Cities are San Fransisco, Los Angeles, and New York , and Boston
# Bottom 4 lowest Cities are Austin, Portalnd, Dallas, and Atlanta.
```

```python
# State Totals (Find Lowest State)
df.groupby("State").sum("Order Total")
StateTotals=df.groupby("State").sum("Order Total")
print(StateTotals)
# Maine, Oregon, and Georgia are the lowest sales states
```

```python
# Regional Totals (Find Lowest Regional Sales)
df.groupby("Region").sum("Order Total")
RegionTotals=df.groupby("Region").sum("Order Total")
print(RegionTotals)
df.groupby("Region").count()["Order Total"].plot(kind='line')
plt.title('Regional Product Sales Total')
# The South is the lowest regional sales. ( The focus for my EDA and I will look into product, sales and time and date trends)
```
